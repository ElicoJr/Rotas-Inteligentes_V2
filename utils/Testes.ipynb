```python
# v2/plot_routes.py
import argparse
from pathlib import Path
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import hashlib, math, requests

# tenta importar base fixa (lon/lat) e OSRM_URL
try:
    from v2 import config
    BASE_LON = float(getattr(config, "BASE_LON", -63.885464691387746))
    BASE_LAT = float(getattr(config, "BASE_LAT", -8.738508095069408))
    OSRM_URL = str(getattr(config, "OSRM_URL", "http://localhost:5000")).rstrip("/")
except Exception:
    BASE_LON = -63.885464691387746
    BASE_LAT = -8.738508095069408
    OSRM_URL = "http://localhost:5000"

PALETTE = [
    "#1f77b4","#ff7f0e","#2ca02c","#d62728","#9467bd",
    "#8c564b","#e377c2","#7f7f7f","#bcbd22","#17becf",
    "#a55194","#393b79","#637939","#8c6d31","#843c39",
    "#cedb9c","#9c9ede","#bd9e39","#ad494a","#6b6ecf"
]

def _team_color(name: str) -> str:
    if not name or pd.isna(name): return "#444444"
    h = int(hashlib.md5(str(name).encode("utf-8")).hexdigest(), 16)
    return PALETTE[h % len(PALETTE)]

def _as_dt(x): return pd.to_datetime(x, errors="coerce")

def _nice_ts(x):
    if pd.isna(x): return "-"
    try: return pd.to_datetime(x).strftime("%Y-%m-%d %H:%M:%S")
    except Exception: return str(x)

def _center_from_df(df):
    if {"longitude","latitude"}.issubset(df.columns) and df["longitude"].notna().any():
        lon = pd.to_numeric(df["longitude"], errors="coerce").dropna()
        lat = pd.to_numeric(df["latitude"],  errors="coerce").dropna()
        if not lon.empty and not lat.empty:
            return float(lat.mean()), float(lon.mean())
    return float(BASE_LAT), float(BASE_LON)

def _load_parquet(args) -> (pd.DataFrame, Path):
    if args.arquivo:
        p = Path(args.arquivo)
        if not p.exists(): raise FileNotFoundError(f"Arquivo não encontrado: {p}")
    else:
        if not args.dia: raise ValueError("Informe --dia (YYYY-MM-DD) ou --arquivo (.parquet).")
        date = pd.to_datetime(args.dia, errors="raise").date()
        p = Path(args.results_dir) / f"atribuicoes_{date}.parquet"
        if not p.exists(): raise FileNotFoundError(f"Arquivo não encontrado: {p}")
    df = pd.read_parquet(p)
    # normaliza campos de data usados na ordenação / popup
    for c in ["datasol","dth_chegada_estimada","dth_final_estimada","fim_turno_estimado"]:
        if c in df.columns: df[c] = _as_dt(df[c])
    return df, p

def _osrm_route_geojson(coords):
    """
    coords: lista [(lon,lat), ...]
    retorna lista de pontos da geometria [(lat,lon), ...] para folium
    """
    if len(coords) < 2:
        return []

    locs = ";".join([f"{lon},{lat}" for (lon,lat) in coords])
    url = f"{OSRM_URL}/route/v1/driving/{locs}?overview=full&geometries=geojson&steps=false"
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    data = r.json()
    routes = data.get("routes", [])
    if not routes: return []
    geom = routes[0].get("geometry", {})
    if geom.get("type") == "LineString":
        # OSRM devolve em [lon,lat]; folium espera [lat,lon]
        return [[lat, lon] for lon, lat in geom.get("coordinates", [])]
    return []

def build_map_with_routes(df: pd.DataFrame, use_cluster: bool = True) -> folium.Map:
    needed = {"longitude","latitude","equipe","numos"}
    miss = [c for c in needed if c not in df.columns]
    if miss:
        raise ValueError(f"Colunas ausentes para plotagem: {miss}")

    center_lat, center_lon = _center_from_df(df)
    m = folium.Map(location=[center_lat, center_lon], zoom_start=12, control_scale=True)

    # marcador da base (ponto único; usamos também para start/end)
    folium.Marker(
        location=[BASE_LAT, BASE_LON],
        icon=folium.Icon(color="blue", icon="home"),
        tooltip="Base (saída/chegada)",
        popup=f"Base fixa<br>lon: {BASE_LON:.6f}<br>lat: {BASE_LAT:.6f}",
    ).add_to(m)

    # ordenar por chegada estimada; se NaT, datasol; se NaT, mantém ordem original
    df = df.copy()
    if "dth_chegada_estimada" not in df.columns:
        df["dth_chegada_estimada"] = pd.NaT

    df["_ord_key"] = df["dth_chegada_estimada"]
    if df["_ord_key"].isna().any() and "datasol" in df.columns:
        # usa datasol como fallback
        ds = _as_dt(df["datasol"])
        df.loc[df["_ord_key"].isna(), "_ord_key"] = ds

    # numerar ordem por equipe
    df["ordem_na_equipe"] = (
        df.sort_values(["equipe","_ord_key"], na_position="last")
          .groupby("equipe").cumcount() + 1
    )

    # um layer por equipe: rota + pontos
    for equipe in sorted(df["equipe"].fillna("N/D").unique()):
        sub = df[df["equipe"] == equipe].copy()
        if sub.empty: continue

        color = _team_color(equipe)
        layer = folium.FeatureGroup(name=f"Equipe {equipe}", show=True).add_to(m)

        # sequência de visita: base → serviços ordenados → base
        sub = sub.sort_values("_ord_key", na_position="last")
        seq = [(BASE_LON, BASE_LAT)]
        for _, r in sub.iterrows():
            lon = pd.to_numeric(r.get("longitude"), errors="coerce")
            lat = pd.to_numeric(r.get("latitude"),  errors="coerce")
            if pd.isna(lon) or pd.isna(lat): continue
            seq.append((float(lon), float(lat)))
        seq.append((BASE_LON, BASE_LAT))

        # desenha rota via OSRM, se possível
        try:
            line = _osrm_route_geojson(seq)
        except Exception:
            line = []

        if line:
            folium.PolyLine(
                line, color=color, weight=4, opacity=0.75, tooltip=f"Rota {equipe}"
            ).add_to(layer)
        else:
            # fallback: linhas retas (haversine visual)
            coords = [[lat, lon] for (lon, lat) in seq]
            folium.PolyLine(
                coords, color=color, weight=2, opacity=0.5, dash_array="5,7",
                tooltip=f"Rota (aprox.) {equipe}"
            ).add_to(layer)

        # marcadores: partida/chegada na base (cores distintas)
        folium.Marker(
            location=[BASE_LAT, BASE_LON],
            icon=folium.Icon(color="green", icon="play"),
            tooltip=f"{equipe}: Partida",
        ).add_to(layer)
        folium.Marker(
            location=[BASE_LAT, BASE_LON],
            icon=folium.Icon(color="red", icon="stop"),
            tooltip=f"{equipe}: Chegada",
        ).add_to(layer)

        # pontos de serviço (cor diferente das bases – usamos a cor da equipe)
        target = MarkerCluster(name=f"{equipe} - serviços", show=True)
        if use_cluster:
            target.add_to(layer)
        else:
            target = layer

        for _, r in sub.iterrows():
            lon = pd.to_numeric(r.get("longitude"), errors="coerce")
            lat = pd.to_numeric(r.get("latitude"),  errors="coerce")
            if pd.isna(lon) or pd.isna(lat): continue

            ordem = int(r.get("ordem_na_equipe", 0)) if pd.notna(r.get("ordem_na_equipe")) else 0
            numos = r.get("numos", "")
            tipo  = r.get("tipo_serv", "")
            datasol = _nice_ts(r.get("datasol"))
            eta = _nice_ts(r.get("dth_chegada_estimada"))
            etd = _nice_ts(r.get("dth_final_estimada"))
            eta_src = r.get("eta_source", "-")

            tooltip = f"#{ordem} • {tipo} • OS {numos}"
            popup_html = f"""
            <b>Equipe:</b> {equipe}<br>
            <b>OS:</b> {numos}<br>
            <b>Tipo:</b> {tipo}<br>
            <b>Data Solicitação:</b> {datasol}<br>
            <b>Chegada Estimada:</b> {eta}<br>
            <b>Término Estimado:</b> {etd}<br>
            <b>Fonte ETA:</b> {eta_src}<br>
            <b>Coord:</b> ({float(lat):.6f}, {float(lon):.6f})
            """

            folium.CircleMarker(
                location=[float(lat), float(lon)],
                radius=6,
                color=color,           # borda = cor da equipe
                fill=True,
                fill_color="#FFFFFF",  # preenchimento branco para diferenciar da rota
                fill_opacity=0.9,
                tooltip=tooltip,
                popup=popup_html,
            ).add_to(target)

            # etiqueta com ordem
            folium.map.Marker(
                [float(lat), float(lon)],
                icon=folium.DivIcon(
                    html=f"""<div style="font-size:10px;color:#000;background:rgba(255,255,255,0.85);
                                     border:1px solid #666;border-radius:3px;padding:0 2px">{ordem}</div>"""
                )
            ).add_to(layer)

    folium.LayerControl(collapsed=False).add_to(m)
    return m

def main():
    ap = argparse.ArgumentParser(description="Plota rotas por equipe (base → serviços → base) a partir de um Parquet.")
    ap.add_argument("--dia", help="Dia no formato YYYY-MM-DD (usa results/atribuicoes_YYYY-MM-DD.parquet)")
    ap.add_argument("--arquivo", help="Caminho para um .parquet específico (substitui --dia)")
    ap.add_argument("--results_dir", default="results", help="Pasta dos arquivos atribuicoes_*.parquet")
    ap.add_argument("--saida", help="HTML de saída (default: results/maps/route_<dia|nome>.html)")
    ap.add_argument("--no-cluster", action="store_true", help="Desliga MarkerCluster dos pontos de serviço")
    args = ap.parse_args()

    df, src_path = _load_parquet(args)
    use_cluster = not args.no_cluster

    m = build_map_with_routes(df, use_cluster=use_cluster)

    out_dir = Path(args.results_dir) / "maps"
    out_dir.mkdir(parents=True, exist_ok=True)
    if args.saida:
        out_file = Path(args.saida)
    else:
        base_name = src_path.stem.replace("atribuicoes_", "route_")
        out_file = out_dir / f"{base_name}.html"

    m.save(str(out_file))
    print(f"✅ Mapa salvo em: {out_file}")


```

```python
main("2023-01-03")
```
